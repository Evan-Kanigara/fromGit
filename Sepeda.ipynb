```python
keluarb=rute.groupby('from_station_name').count()
keluarb1=keluarb.reset_index()
keluarb11=keluarb1.rename(columns={"from_station_name": "stasiun", "to_station_name": "bicycle_out"})
keluarb11
```

```python
masukb=rute.groupby('to_station_name').count()
masukb1=masukb.reset_index()
masukb11=masukb1.rename(columns={"to_station_name": "stasiun", "from_station_name": "bicycle_in"})
masukb11
```

```python
merge = pd.merge(masukb11, keluarb11)
merge
```

```python
mergef=pd.merge(masukb11, keluarb11,how='outer')
mergef
```
